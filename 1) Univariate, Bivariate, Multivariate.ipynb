## Univariate Analysis
It is a type of data visualization where we visualize only a single variable at a time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('employee.csv')

data.shape
```

```python
data.head()
```

### Univariate Analysis on numerical variables

```python
sns.distplot(data['Age'])
```

### Univariate Analysis on Categorical Variables

```python
sns.countplot(data['BusinessTravel'])
```

## Bivariate Analysis

- There are 2 types of Bivariate:

1) Categorical vs Numerical
2) Numerical vs Numerical

### Categorical vs Numerical

```python
sns.barplot(data['Department'], data['MonthlyIncome'])
```

### Continuous VS Continuous

```python
sns.scatterplot(data['Age'], data['MonthlyIncome'])
```

## Multivariate Analysis


## heat Map

```python
plt.rcParams['figure.figsize'] = (19,8)
sns.heatmap(data.corr(), annot=True, fmt='0.1f')
plt.show()

# annot--->means to see the value also
# fmt---->we have to see co-relation value with one decimal place
```

### Barplot with extra variable

```python
sns.barplot(data['Department'], data['MonthlyIncome'], data['Attrition'])
```
